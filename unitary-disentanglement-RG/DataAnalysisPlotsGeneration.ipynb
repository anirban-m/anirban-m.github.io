```python
#from ipynb.fs.full.CouplingRGFuncs import LambdasForEntanglementRG
#lambdas=LambdasForEntanglementRG()
#print(lambdas)
#Designing the momentum space coordinate system
#LabelCoord=[[]]
#0 1
def physicalDistance(chklabel):
    #states along the normal to kF1=\pi/2,\pi/2
    kF1=np.array([np.pi/2,np.pi/2])
    kx1=kF1[0]-lambdas
    ky1=kF1[1]-lambdas
    k1=np.dstack([kx1,ky1])[0]
    labels1=np.array([0,1,2,3,4,5,6])
    #states along the normal to kF1=\pi/4,3\pi/4
    kF2=np.array([np.pi/4,3*np.pi/4])
    kx2=kF2[0]-lambdas
    ky2=kF2[1]-lambdas
    k2=np.dstack([kx2,ky2])[0]
    labels2=np.array([7,8,9,10,11,12,13])
    #states along the normal to kF1=-\pi/2,-\pi/2
    kF3=np.array([-np.pi/2,-np.pi/2])
    kx3=kF3[0]+lambdas
    ky3=kF3[1]+lambdas
    k3=np.dstack([kx3,ky3])[0]
    labels3=np.array([14,15,16,17,18,19,20])
    #states along the normal to kF4=-\pi/4,-3\pi/4
    kF4=np.array([-np.pi/4,-3*np.pi/4])
    kx4=kF4[0]+lambdas
    ky4=kF4[1]+lambdas
    k4=np.dstack([kx4,ky4])[0]
    labels4=np.array([21,22,23,24,25,26,27])
    labels=np.stack([labels1,labels2,labels3,labels4])
    kx=np.stack([kx1,kx2,kx3,kx4])
    ky=np.stack([ky1,ky2,ky3,ky4])
    indexLabel=np.where(labels==chklabel[0])
    kx_1=kx[indexLabel[0][0]][indexLabel[1][0]]
    ky_1=ky[indexLabel[0][0]][indexLabel[1][0]]
    indexLabel=np.where(labels==chklabel[1])
    kx_2=kx[indexLabel[0][0]][indexLabel[1][0]]
    ky_2=ky[indexLabel[0][0]][indexLabel[1][0]]
    return 1/(abs(kx_1-kx_2)+abs(ky_1-ky_2))
geopair0=np.asarray(pairs0)[np.where(dists0==g0)][1]
geopair1=np.asarray(pairs1)[np.where(dists1==g1)][0]
geopair2=np.asarray(pairs2)[np.where(dists2==g2)][0]
geopair3=np.asarray(pairs3)[np.where(dists3==g3)][0]
geopair4=np.asarray(pairs4)[np.where(dists4==g4)][0]
geopair5=np.asarray(pairs5)[np.where(dists5==g5)][0]
geopair6=np.asarray(pairs6)[np.where(dists6==g6)][0]
#geoPoints=np.asarray([','.join(geopair0.astype(str)),geopair1,geopair2,geopair3,geopair4,geopair5,geopair6])
geopairs=np.array([geopair0,geopair1,geopair2,geopair3,geopair4,geopair5,geopair6])
geoPoints=[','.join(eval('geopair%s'%i).astype(str)) for i in range(7)]
count=0
posArr=np.array([(25,12),(25,10),(25,10),(35,0),(-10, -15),(13, -25),(13, -25)])
dpPairs=np.asarray([physicalDistance(pair)*2*np.pi for pair in geopairs])
fig,ax=plt.subplots(1,2)
plt.rcParams['text.usetex'] = True
plt.rcParams['figure.figsize'] = [15,6]
ax[0].plot(np.array([g0,g1,g2,g3,g4,g5,g6]),marker='o',linestyle='--',color='red')#,geodesic5
ax[1].plot(dpPairs,linestyle='--',marker='o')
ax[0].set_ylabel(r"$d_{g}$",fontsize=20)
ax[0].set_xlabel(r'\Huge{$\longleftarrow$}~~Steps',fontsize=20)
ax[0].tick_params(labelsize=15)
ax[1].tick_params(labelsize=15)
ax[1].set_ylabel(r"$d_{p}$(in units of a)",fontsize=20)
ax[1].set_xlabel(r'\Huge{$\longleftarrow$}~~Steps',fontsize=20)
for label, x, y in zip(geoPoints, np.arange(7), dpPairs):
    ax[1].annotate(
        label,
        xy = (x, y), xytext = posArr[count],
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.3', fc = 'yellow', alpha = 0.7),size=14)
        #arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'),
        #weight='bold')
    count=count+1
    
    
    #plt.rcParams['figure.figsize'] = [15,6]
    #plt.rcParams['text.usetex'] = True
    #plt.xlabel(r'\Huge{$\longleftarrow$}~~Steps',fontsize=20)
```

Given the feature learning ability of the RG seen above we ask the following question:<br>
Can the Entanglement Holographic Mapping network function as a Deep neural network? <br>

Predictive abilities of this RG can be checked in the following way: given a pair(i,j) at each RG step we compute the conditional probability P(S|(i,j)) of it being strong entangled pair finally? 
Below will layout the steps neccessary to answer this question.<br>
1>Take as input the pair in question and (Total-N(No of RG transformation steps performed))<br>
2>Check if the ordered pair entities belong to entangled subspace after N transformation layers:
if answer is No then return P(S|(i,j))=0(as it has been disentangled at that or earlier layer) <br>
else follow the steps below<br>
&nbsp;&nbsp;&nbsp;&nbsp;1>extract the mutual information(MI) features for every pair from the output of the Nth transformation layer.<br>
&nbsp;&nbsp;&nbsp;&nbsp;2>Define a discretizer that creates a integer feature set with bin values [0,1,...,Num=int(2log2/min(mutInf among all layers))]<br>
&nbsp;&nbsp;&nbsp;&nbsp;3>Define a classifier that labels all features with value =<Num/2 as 'weak':=0 and value>Num/2 as 'strong':=1.<br>
&nbsp;&nbsp;&nbsp;&nbsp;4>Obtain the classification(pairClass0) using(steps 2 and 3) of the features with respect to the target feature set(at RG fixed point).<br>
&nbsp;&nbsp;&nbsp;&nbsp;5>Obtain the discretized feature set pairFeatN using step 2 for the MI features from the output of the Nth transformation layer.<br>
&nbsp;&nbsp;&nbsp;&nbsp;6>Compute the Joint probability distribution between pairFeatN and pairClass0, This tell us the association probability between the target class and<br>
&nbsp;&nbsp;&nbsp;&nbsp;bottleneck representation of features after N transformations of initial layer.
