$|\Psi\rangle=|0_{i}0_{j}\rangle|\Psi_{00}\rangle+|0_{i}1_{j}\rangle|\Psi_{01}\rangle+|1_{i}0_{j}\rangle|\Psi_{10}\rangle+|1_{i}1_{j}\rangle|\Psi_{11}\rangle$<br>
$|\Psi\rangle=|0\rangle|\Psi^{\perp}_{0}\rangle+|1\rangle|\Psi_{1}\rangle+|2\rangle|\Psi_{2}\rangle+|3\rangle|\Psi_{3}\rangle$<br>
$|\Psi\rangle=|0\rangle|\Psi^{\perp}_{0}\rangle+|1\rangle|\Psi^{\perp}_{1}\rangle+c_{01}|1\rangle|\Psi_{0}^{\perp}\rangle+|2\rangle|\Psi_{2}^{\perp}\rangle+c_{12}|2\rangle|\Psi_{1}^{\perp}\rangle+c_{02}|2\rangle|\Psi^{\perp}_{0}\rangle+|3\rangle\Psi_{3}^{\perp}\rangle + c_{23}|3\rangle|\Psi^{\perp}_{2}\rangle+c_{13}|3\rangle|\Psi^{\perp}_{1}\rangle+c_{03}|3\rangle|\Psi^{\perp}_{0}\rangle$<br>
# Gram Schmidt Decomposition<br>
$00\equiv 0$, $01\equiv 1$, $10\equiv 2$, $11\equiv 3$,$c_{ij}=\dfrac{\langle\Psi_{i}^{\perp}|\Psi_{j}\rangle}{\langle\Psi_{i}^{\perp}|\Psi_{i}^{\perp}\rangle}$<br>
__Step-0:-__$|\Psi_{0}^{\perp}\rangle=|\Psi_{0}\rangle$<br>
__Step-1:-__ $|\Psi^{\perp}_{1}\rangle= |\Psi_{1}\rangle-c_{01}|\Psi^{\perp}_{0}\rangle$, $\langle\Psi_{0}|\Psi^{\perp}_{1}\rangle = 0$<br>
__Step-2:-__ $|\Psi^{\perp}_{2}\rangle= |\Psi_{2}\rangle-c_{12}|\Psi^{\perp}_{1}\rangle-c_{02}|\Psi^{\perp}_{0}\rangle$, $\langle\Psi_{1}^{\perp}|\Psi^{\perp}_{2}\rangle=0, \langle\Psi_{0}|\Psi^{\perp}_{2}\rangle=0$<br>
__Step-3:-__ $|\Psi^{\perp}_{3}\rangle= |\Psi_{3}\rangle-c_{23}|\Psi^{\perp}_{2}\rangle-c_{13}|\Psi^{\perp}_{1}\rangle-c_{03}|\Psi^{\perp}_{0}\rangle$, $\langle\Psi_{01}^{\perp}|\Psi^{\perp}_{11}\rangle=0, \langle\Psi_{00}|\Psi^{\perp}_{11}\rangle=0$, $\langle\Psi_{00}|\Psi^{\perp}_{11}\rangle=0$<br>

__Step-4:-__$|\Psi\rangle = |\Psi\rangle=|0\rangle|\Psi^{\perp}_{0}\rangle+|1\rangle|\Psi^{\perp}_{1}\rangle+c_{01}|1\rangle|\Psi_{0}^{\perp}\rangle+|2\rangle|\Psi_{2}^{\perp}\rangle+c_{12}|2\rangle|\Psi_{1}^{\perp}\rangle+c_{02}|2\rangle|\Psi^{\perp}_{0}\rangle+|3\rangle\Psi_{3}^{\perp}\rangle + c_{23}|3\rangle|\Psi^{\perp}_{2}\rangle+c_{13}|3\rangle|\Psi^{\perp}_{1}\rangle+c_{03}|3\rangle|\Psi^{\perp}_{0}\rangle$

\begin{align}
|\Psi\rangle = \begin{pmatrix}|\phi^{\perp}_{0}\rangle & |\phi^{\perp}_{1}\rangle & |\phi^{\perp}_{2}\rangle & |\phi^{\perp}_{3}\rangle\end{pmatrix}D\begin{pmatrix}|0\rangle \\ |1\rangle \\ |2\rangle \\ |3\rangle\end{pmatrix}, D=\frac{1}{\sqrt{\sum_{ij}\bar{c}_{ij}^{2}}}\begin{pmatrix} 
\bar{c}_{00} & \bar{c}_{01} & \bar{c}_{02} & \bar{c}_{03}\\ 
0      & \bar{c}_{11} & \bar{c}_{12} & \bar{c}_{13}\\
0      & 0      & \bar{c}_{22} & \bar{c}_{23}\\
0      & 0      & 0      & \bar{c}_{33}
\end{pmatrix}\end{align}, where $|\phi_{i}^{\perp}\rangle =\frac{1}{\sqrt{\langle\Psi^{\perp}_{i}|\Psi^{\perp}_{i}}\rangle}|\Psi^{\perp}_{i}\rangle$ and $\bar{c}_{ij}=c_{ij}\sqrt{c_{ii}}(1-\delta_{ij})+\delta_{ij}\sqrt{c_{ii}}$<br>
$|\Psi\rangle=\sum_{ij}D_{ij}|i\rangle|\phi^{\perp}_{j}\rangle=\sum\limits^{3}_{j=0}|0\rangle\left\lbrace D_{0j}|0\rangle+D_{1j}|1\rangle\right\rbrace|\phi^{\perp}_{j}\rangle+|1\rangle\left\lbrace D_{2j}|0\rangle+D_{3j}|1\rangle\right\rbrace|\phi^{\perp}_{j}\rangle$


Decomposition of $|\Psi\rangle$ w.r.t four configurations of pseudospins i,j and the rest subsystem. The subsystem states being  $|\Psi_{00}\rangle$, $|\Psi_{01}\rangle$, $|\Psi_{10}\rangle$, $|\Psi_{11}\rangle$ with $\langle\Psi_{a}|\Psi_{b}\rangle \neq 0$ where $a,b=00,10,01,11$. 
$|\Psi\rangle=|0_{i}0_{j}\rangle|\Psi_{00}\rangle+|0_{i}1_{j}\rangle|\Psi_{01}\rangle+|1_{i}0_{j}\rangle|\Psi_{10}\rangle+|1_{i}1_{j}\rangle|\Psi_{11}\rangle$. 
The Schmidt decomposition for this state happens as follows:
$|\Psi\rangle=c_{0}|\psi^{0}_{ij}\rangle|\phi_{0}\rangle+c_{1}|\psi^{1}_{ij}\rangle|\phi_{1}\rangle+c_{2}|\psi^{2}_{ij}\rangle|\phi_{2}\rangle+c_{3}|\psi^{3}_{ij}\rangle|\phi_{3}\rangle$ where $\langle\psi^{m}_{ij}|\psi^{n}_{ij}\rangle=\delta_{mn}$ and $\langle\phi_{m}|\phi_{n}\rangle=\delta_{mn}$ 


```python
import numpy as np
import re
from numpy import linalg as LA
from time import time
from scipy import sparse
def InnerProd(A,B):
    hashedA = np.array([hash(s) for s in A['keys']])
    hashedB = np.array([hash(s) for s in B['keys']])
    AdjMat=np.equal.outer(hashedB, hashedA)
    SparsedAdjMat=sparse.csr_matrix(AdjMat)
    IP=np.sum(sparse.csr_matrix.dot(sparse.csr_matrix.dot(B['vals'],SparsedAdjMat),A['vals']))
    return IP
#def InnerProd(Psi0Rec,Psi1Rec):
#    InnerProd=0
    #Psi0Rec = np.rec.fromarrays([Psi0[:,0],Psi0[:,1].astype(np.float64)], names='keys,vals')
    #Psi1Rec = np.rec.fromarrays([Psi1[:,0],Psi1[:,1].astype(np.float64)], names='keys,vals')
#    for i in range(len(Psi0Rec)):
#        InnerProd=InnerProd+ np.sum(Psi0Rec['vals'][i]*(Psi1Rec['keys']==Psi0Rec['keys'][i])*Psi1Rec['vals'])#np.sum(().astype(np.float64)*((Psi1Rec['keys']==conf)*))
#    return InnerProd 
```

```python
def GSOrth(PsiOld,PsiNonOrth):
    innerProd=InnerProd(PsiOld,PsiNonOrth)
    norm=InnerProd(PsiOld,PsiOld)
    if(innerProd==0):
        return PsiNonOrth
    else:
        supCoeff=(innerProd/norm)
        PsiUnionKeys=np.union1d(PsiOld['keys'],PsiNonOrth['keys'])
        PsiCompose=np.array([0.+np.sum(PsiNonOrth['vals'][PsiNonOrth['keys']==config])-supCoeff*np.sum(PsiOld['vals'][PsiOld['keys']==config]) for config in PsiUnionKeys],dtype='float')
        PsiCompose1=PsiCompose[PsiCompose.astype(float)!=0.0]
        PsiUnionKeys=PsiUnionKeys[PsiCompose.astype(float)!=0.0]
        PsiNew=np.rec.fromarrays([PsiUnionKeys,PsiCompose1],names='keys,vals')
        return PsiNew
```

```python
def SchmidtCoeffBiPartPairFromRest(i,j):
    if(i!=j):
        #print("start",i,j)
        #manyBodyKet=np.array([np.array([np.array([eval(qbit) for qbit in list(conf[:N])]),eval(conf[N:])]) for conf in PsiConf[2:]])
        manyBodyKet=np.array([np.array([conf[:N],eval(conf[N:])]) for conf in PsiConf[2:]],dtype='<U64')
        Ket=np.rec.fromarrays([manyBodyKet[:,0],manyBodyKet[:,1].astype(float)],dtype=[('keys','<U64'),('vals','>f8')])
        BitConfiGij00=''.join(['0' if((ind==i) or (ind==j)) else '.' for ind in range(N)])
        BitConfiGij01=''.join(['0' if(ind==i) else '1' if(ind==j) else '.' for ind in range(N)])
        BitConfiGij10=''.join(['1' if(ind==i) else '0' if(ind==j) else '.' for ind in range(N)])
        BitConfiGij11=''.join(['1' if((ind==i) or (ind==j)) else '.' for ind in range(N)])
        r00 = re.compile(BitConfiGij00)
        r01 = re.compile(BitConfiGij01)
        r10 = re.compile(BitConfiGij10)
        r11 = re.compile(BitConfiGij11)
        OrdInd=(j,i) if i>j else (i,j)
        PsiSubSystem0 = ['00',Ket[np.vectorize(lambda x:bool(r00.match(x)))(Ket['keys'])]]
        PsiSubSystem1 = ['01',Ket[np.vectorize(lambda x:bool(r01.match(x)))(Ket['keys'])]]
        PsiSubSystem2 = ['10',Ket[np.vectorize(lambda x:bool(r10.match(x)))(Ket['keys'])]]
        PsiSubSystem3 = ['11',Ket[np.vectorize(lambda x:bool(r11.match(x)))(Ket['keys'])]]
        PsiSubSystemConfs=[]
        for a in range(4):
            if((0 in eval('PsiSubSystem'+str(a))[1].shape)==False):
                tag=eval('PsiSubSystem'+str(a))[0]
                Psi=eval('PsiSubSystem'+str(a))[1]
                for conf in Psi['keys']:
                    Psi['keys'][Psi['keys']==conf]=Psi['keys'][Psi['keys']==conf][0][0:OrdInd[0]]+Psi['keys'][Psi['keys']==conf][0][OrdInd[0]+1:OrdInd[1]]+Psi['keys'][Psi['keys']==conf][0][OrdInd[1]+1:N]
                sqrtIP=np.sqrt(InnerProd(Psi,Psi))
                Psi['vals']=(1/sqrtIP)*Psi['vals']
                PsiSubSystemConfs=PsiSubSystemConfs+[[tag,Psi,sqrtIP]]
        #print("old")
        #for psi in PsiSubSystemConfs:
        #    print(psi)
        PsiSubSystemConfsCopy=[PsiSubSystemConfs[0]]
        for i in range(len(PsiSubSystemConfs)-1):
            num=len(PsiSubSystemConfsCopy)
            cond=True
            sign=1
            for j in range(len(PsiSubSystemConfsCopy)):
                    IP=InnerProd(PsiSubSystemConfsCopy[j][1],PsiSubSystemConfs[i+1][1])
                    if(abs(IP+1)<1e-8):
                        cond=False
                        sign=-1
            if(cond is False):        
                PsiSubSystemConfsCopy=PsiSubSystemConfsCopy+[[PsiSubSystemConfs[i+1][0],PsiSubSystemConfsCopy[j][1],sign*PsiSubSystemConfs[i+1][2]]]
            else:
                PsiSubSystemConfsCopy=PsiSubSystemConfsCopy+[PsiSubSystemConfs[i+1]]                                           
        PsiSubSystemConfs=PsiSubSystemConfsCopy
        #unique two pseudospin configurations
        uniqueTags=np.unique([conf[0] for conf in PsiSubSystemConfs])
        print(uniqueTags)
        #print("new",PsiSubSystemConfs)
        PsiUniqueConfs=[PsiSubSystemConfs[0][1]]
        for i in range(len(PsiSubSystemConfs)):
            cond=True
            for j in range(len(PsiUniqueConfs)):
                if(abs(InnerProd(PsiSubSystemConfs[i][1],PsiUniqueConfs[j])-1)<1e-5):
                    cond=False       
            if (cond is True):
                PsiUniqueConfs=PsiUniqueConfs+[PsiSubSystemConfs[i][1]]
        #print("new")
        #for psi in PsiSubSystemConfs:
        #    print(psi)
        #print("****Preparing Orthonormal basis****")
        NonOrthoNum=len(PsiUniqueConfs)
    #Gram Schmidt Orthogonalization of the subsystem states
    ################################################################################
        if(NonOrthoNum==1):
            #print('a')
        #In the case when there is only one subsystem state the array below stores one state
            PsiOrtho=[PsiUniqueConfs[0]]
        elif(NonOrthoNum==2):
            #print('b')
            PsiOrtho=[PsiUniqueConfs[0]]
            Psi1=GSOrth(PsiOrtho[0],PsiUniqueConfs[1])
            IP=InnerProd(Psi1,PsiOrtho[0])
            if(abs(IP)<1e-7):
                #print("Orthogonalization Successful")
                #normalizing the Orthogonalized State
                Psi1['vals']=(1/np.sqrt(InnerProd(Psi1,Psi1)))*Psi1['vals']
                if abs(InnerProd(Psi1,Psi1)-1)<1e-5:
                    #print("normalization Successful")
                    PsiOrtho=PsiOrtho+[Psi1]
            else:
                print("FAIL-1")
                return "FAILED"
        elif(NonOrthoNum==3):
            #print('c')
            PsiOrtho=[PsiUniqueConfs[0]]
            Psi1=GSOrth(PsiOrtho[0],PsiUniqueConfs[1])
            IP=InnerProd(Psi1,PsiOrtho[0])
            if(abs(IP)<1e-7):
                #print("Orthogonalization Successful")
                #normalizing the Orthogonalized State
                Psi1['vals']=(1/np.sqrt(InnerProd(Psi1,Psi1)))*Psi1['vals']
                if abs(InnerProd(Psi1,Psi1)-1)<1e-5:
                    #print("Orthonormalization complete")
                    PsiOrtho=PsiOrtho+[Psi1]
            else:
                print("FAIL-1")
                return "FAILED"
            Psi2=GSOrth(PsiOrtho[1],PsiUniqueConfs[2])
            IP=InnerProd(Psi2,PsiOrtho[1])
            if(abs(IP)<1e-7):
                Psi2=GSOrth(PsiOrtho[0],Psi2)
                IP=InnerProd(Psi2,PsiOrtho[0])
                if(abs(IP)<1e-7):
                    #print("Orthogonalization Successful")
                    #normalizing the Orthogonalized State
                    Psi2['vals']=(1/np.sqrt(InnerProd(Psi2,Psi2)))*Psi2['vals']
                    if abs(InnerProd(Psi2,Psi2)-1)<1e-5:
                        #print("Orthonormalization complete")
                        PsiOrtho=PsiOrtho+[Psi2]
                else:
                    print("FAIL-3")
                    return "FAILED"
            else:
                print("FAIL-2")
                return "FAILED"
        elif(NonOrthoNum==4):
            #print('d')
            PsiOrtho=[PsiUniqueConfs[0]]
            Psi1=GSOrth(PsiOrtho[0],PsiUniqueConfs[1])           
            IP=InnerProd(Psi1,PsiOrtho[0])
            if(abs(IP)<1e-7):    
                #print("Orthogonalization Successful")
                #normalizing the Orthogonalized State
                Psi1['vals']=(1/np.sqrt(InnerProd(Psi1,Psi1)))*Psi1['vals']
                if abs(InnerProd(Psi1,Psi1)-1)<1e-5:
                    #print("Orthonormalization complete")
                    PsiOrtho=PsiOrtho+[Psi1]
                else:
                    print("FAIL-1")
                    return "FAILED"    
            Psi2=GSOrth(PsiOrtho[1],PsiUniqueConfs[2])
            IP=InnerProd(Psi2,PsiOrtho[1])
            if(abs(IP)<1e-7):
                Psi2=GSOrth(PsiOrtho[0],Psi2)
                IP=InnerProd(Psi2,PsiOrtho[0])
                if(abs(IP)<1e-7):
                    #print("Orthogonalization Successful")
                    #normalizing the Orthogonalized State
                    Psi2['vals']=(1/np.sqrt(InnerProd(Psi2,Psi2)))*Psi2['vals']
                    if abs(InnerProd(Psi2,Psi2)-1)<1e-5:
                        #print("Orthonormalization complete")
                        PsiOrtho=PsiOrtho+[Psi2]
                else:
                    print("FAIL-3")
                    return "FAILED"
            else:
                print("FAIL-2")
                return "FAILED"
            Psi3=GSOrth(PsiOrtho[2],PsiUniqueConfs[3])
            IP=InnerProd(Psi3,PsiOrtho[2])
            if(abs(IP)<1e-7):
                Psi3=GSOrth(PsiOrtho[1],Psi3)
                IP=InnerProd(Psi3,PsiOrtho[1])
                if(abs(IP)<1e-7):
                    Psi3=GSOrth(PsiOrtho[0],Psi3)
                    IP=InnerProd(Psi3,PsiOrtho[0])
                    if(abs(IP)<1e-7):
                        #print("Orthogonalization Successful")
                        #normalizing the Orthogonalized State
                        Psi3['vals']=(1/np.sqrt(InnerProd(Psi3,Psi3)))*Psi3['vals']
                        if abs(InnerProd(Psi3,Psi3)-1)<1e-5:
                            #print("Orthonormalization complete")
                            PsiOrtho=PsiOrtho+[Psi3]
                    else:
                        print("FAIL-6")
                        return "FAILED"
                else:
                    print("FAIL-5")
                    return "FAILED"
            else:
                print("FAIL-4")
                return "FAILED"
        #print("Checking integrity of Orthonormal basis")
        boolCond=True
        for i in range(len(PsiOrtho)):
            for j in range(len(PsiOrtho)):
                if(i!=j):
                    if(abs(InnerProd(PsiOrtho[i],PsiOrtho[j]))>1e-7):
                        boolCond=False
                else:
                    if(abs(InnerProd(PsiOrtho[i],PsiOrtho[j])-1)>1e-7):
                        boolCond=False
        #if(boolCond==True):
        #    print("Orthonormal basis formed")
        if(boolCond==False):
            print("FAILED")
            return ''  
        ###########################################
                #Computing the D matrix
        ############################################
        D=np.zeros([len(uniqueTags),len(PsiOrtho)])
        #extracting the tags associated with the two spin configurations 
        #within the members in Psi decomposed into subsystems->PsiSubSystemConfs
        tags=[psi[0] for psi in PsiSubSystemConfs]
        for i in range(len(uniqueTags)):
            for j in range(len(PsiOrtho)):
                ind=np.where(np.array(tags)==uniqueTags[i])[0]
                Sum=0
                for index in range(len(ind)):
                    Sum=Sum+InnerProd(PsiSubSystemConfs[ind[index]][1],PsiOrtho[j])*PsiSubSystemConfs[ind[index]][2]
                D[i][j]=Sum
        #Checking normalization from D matrix
        norm=np.sum(D.flatten()*D.flatten())
        #if(np.abs(norm-1)<1e-7):
        #    print("D matrix constructed succesfully")
        if(np.abs(norm-1)>1e-7):
            print("D matrix construction failed")
            return "FAILED"
        u,SchmidtCoeffs,v=np.linalg.svd(D, full_matrices=True)
        #obtaining the 1-pseudospin reduced density matrices from the 2-pseudospin reduced density matrices
        #This leads to the two particle reduced density matrix
        Rho_ij=np.dot(D,D.T)
        #configs of pseudospin-1
        confOf1=np.asarray([np.array([tag[0],tag[1]]) for tag in uniqueTags])
        #building diagonal entries for 1-pseudospin reduced density matrix--ith Spin
        iRho00=0 if 0 in np.where((confOf1[:,0]=='0'))[0].shape else sum([Rho_ij[ind][ind] for ind in np.where((confOf1[:,0]=='0'))[0]])
        #building offdiagonal entries for 1-pseudospin reduced density matrix
        ind00= np.where((uniqueTags=='00'))[0]
        ind01= np.where((uniqueTags=='01'))[0]
        ind10= np.where((uniqueTags=='10'))[0]
        ind11= np.where((uniqueTags=='11'))[0]
        if((0 in ind00.shape) | (0 in ind10.shape)):
            comp1=0
        else:
            comp1=Rho_ij[ind00[0]][ind10[0]]
        if((0 in ind01.shape) | (0 in ind11.shape)):
            comp2=0
        else:
            comp2=Rho_ij[ind01[0]][ind11[0]]
        iRho01=comp1+comp2
        #reduced density Matrix for i Pseudospin
        iRho=np.array([np.array([iRho00,iRho01]),np.array([iRho01,1-iRho00])])
        
        #building diagonal entries for 1-pseudospin reduced density matrix--jth Spin
        jRho00=0 if 0 in np.where((confOf1[:,1]=='0'))[0].shape else sum([Rho_ij[ind][ind] for ind in np.where((confOf1[:,1]=='0'))[0]])
        #building offdiagonal entries for 1-pseudospin reduced density matrix
        if((0 in ind00.shape) | (0 in ind01.shape)):
            comp1=0
        else:
            comp1=Rho_ij[ind00[0]][ind01[0]]
        if((0 in ind10.shape) | (0 in ind11.shape)):
            comp2=0
        else:
            comp2=Rho_ij[ind10[0]][ind11[0]]
        jRho01=comp1+comp2
        #reduced density Matrix for j Pseudospin
        jRho=np.array([np.array([jRho00,jRho01]),np.array([jRho01,1-jRho00])])
        #eigenvalues of the 1-particle reduced density matrix -> eigenvalues are the probabilities
        iRhoDiag=LA.eigvals(iRho)
        jRhoDiag=LA.eigvals(jRho)
        #Probability for ith pseudospin
        p_i=1 if np.abs(iRhoDiag[0]-1)<1e-7 else iRhoDiag[0]
        #Probability for jth pseudospin
        p_j=jRhoDiag[0] 
        #joint probabilities for ijth pseudospin
        p_ij=SchmidtCoeffs*SchmidtCoeffs
        p_ij[np.abs(p_ij-1)<1e-7]=1
        #ith pseudospin entropy
        S_i=-p_i*np.log(p_i)-(1-p_i)*np.log(1-p_i) if p_i*(1-p_i)!=0 else 0
        #jth pseudospin entropy
        S_j=-p_j*np.log(p_j)-(1-p_j)*np.log(1-p_j) if p_j*(1-p_j)!=0 else 0
        #joint entropic weights
        S_ijArr=-p_ij*np.log(p_ij)
        S_ijArr[p_ij*(1-p_ij)==0]=0.0
        S_ij=np.sum(S_ijArr)
        #Mutual Information
        MutInf_ij=S_i+S_j-S_ij
        if(MutInf_ij<-1e-5):
            return ['FAIL',D]
            
        return [p_i,p_j,SchmidtCoeffs,MutInf_ij,-np.log(MutInf_ij/(2*np.log(2)))]  
        ###################################################################################
    else:
        return ''
```

### Distance Map for ground state wavefunction at RG fixed point

```python
arrPoints = np.asarray(points0)
DistMat0=np.zeros([len(arrPoints),len(arrPoints)])
count=0
for i in range(len(arrPoints)):
    for j in range(len(arrPoints)):
        if(i>j):
            DistMat0[i][j]=dists0[count]
            count=count+1
#combining distances of (j,i) coords with (i,j)
DistMat0=DistMat0+DistMat0.T
DistMat0[DistMat0==np.inf]=1e100
```

```python
#Multidimensional scaling using sklearn
mds = manifold.MDS(n_components=2, dissimilarity="precomputed", random_state=6,metric=True)
results= mds.fit(DistMat0)
```

```python
coords = results.embedding_
plt.rcParams['figure.figsize'] = [15,7]
figure,ax=plt.subplots(1,1)
ax.scatter(coords[:, 0], coords[:, 1], marker = 'o', s=120)
count=0
for label, x, y in zip(arrPoints, coords[:, 0], coords[:, 1]):
    ax.annotate(
        label,
        xy = (x, y), xytext = (count+20, -count+20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.3', fc = 'yellow', alpha = 0.7),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'),
        weight='bold')
    count=count+9
# Plot the edges
start_idx, end_idx = np.where(coords)
segments = [[coords[i, :], coords[j, :]]
            for i in range(len(coords)) for j in range(len(coords))]
plt.show()
```

```python
hist,bins=np.histogram(MI0,bins='doane',range=(0,np.max(MI0)),density=False) 
plt.hist(MI0,bins,weights=np.ones(len(MI0))/len(MI0))
plt.xticks(bins)
plt.yticks(np.linspace(0,np.max(hist),np.max(hist)+1)[::2]/len(MI0))
plt.xlim(min(bins),max(bins))
```
